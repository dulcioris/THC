1.- Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
#definimos la funcion
def esprimo(numero):
    if numero <=1:
        return "Error en el Número"
    for i in range(2, int(numero**0.5)+1):
      if numero %i == 0:
        return False
    return True
```

```python
#Script
#creamos la lista
L = list(range(2, 1001))

#filtrado de numeros primos
P = list(filter(esprimo, L))

#Mostramos los elementos en P
for i, primo in enumerate(P, start=1):
  print(f'P({i}) = {primo}')
```

```python
while True:
  try:
    num=int(input("Ingrese un número entero mayor que 1:"))
    resultado= esprimo(num)
    print(resultado)
    break
  except ValueError:
    print("Por favor, ingrese un número válido")
```

2.- Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

*   El promedio
*   La varianza
*   La desviación estándar
*   Visualice los datos en un histograma.
*   La suma de los números pares e impares.
*   Los múltiplos de 2, 3 y 5.

```python
#EL PROMEDIO
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Calcular el promedio utilizando funciones estándar
promedio_estandar = sum(numeros) / len(numeros)

# Calcular el promedio utilizando una función lambda
promedio_lambda = (lambda x: sum(x) / len(x))(numeros)

print("Promedio (función estándar):", promedio_estandar)
print("Promedio (función lambda):", promedio_lambda)
```

```python
#LA VARIANZA
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Calcular el promedio
promedio = sum(numeros) / len(numeros)

# Calcular la varianza utilizando funciones estándar
varianza_estandar = sum((x - promedio) ** 2 for x in numeros) / len(numeros)

# Calcular la varianza utilizando una función lambda
varianza_lambda = (lambda x, m: sum((i - m) ** 2 for i in x) / len(x))(numeros, promedio)

print("Varianza (función estándar):", varianza_estandar)
print("Varianza (función lambda):", varianza_lambda)
```

```python
#LA DESVIACIÓN ESTÁNDAR
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Calcular el promedio
promedio = sum(numeros) / len(numeros)

# Calcular la varianza
varianza = sum((x - promedio) ** 2 for x in numeros) / len(numeros)

# Calcular la desviación estándar utilizando funciones estándar
desviacion_estandar = varianza ** 0.5

# Calcular la desviación estándar utilizando una función lambda
desviacion_lambda = (lambda x, m, v: v ** 0.5)(numeros, promedio, varianza)

print("Desviación estándar (función estándar):", desviacion_estandar)
print("Desviación estándar (función lambda):", desviacion_lambda)
```

```python
#VISUALIZACIÓN
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Función para imprimir el histograma
def visualizar_histograma(datos):
    for i in range(max(datos) + 1):
        count = datos.count(i)
        print(f"{i}: {'*' * count}")

# Visualizar histograma utilizando funciones estándar
visualizar_histograma(numeros)
```

```python
#SUMA DE NÚMEROS PARES E IMPARES
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Calcular la suma de números pares e impares utilizando funciones estándar
suma_pares_estandar = sum(x for x in numeros if x % 2 == 0)
suma_impares_estandar = sum(x for x in numeros if x % 2 != 0)

# Calcular la suma de números pares e impares utilizando una función lambda
suma_pares_lambda = (lambda x: sum(i for i in x if i % 2 == 0))(numeros)
suma_impares_lambda = (lambda x: sum(i for i in x if i % 2 != 0))(numeros)

print("Suma de números pares (función estándar):", suma_pares_estandar)
print("Suma de números impares (función estándar):", suma_impares_estandar)

print("Suma de números pares (función lambda):", suma_pares_lambda)
print("Suma de números impares (función lambda):", suma_impares_lambda)
```

```python
#MÚLTIPLOS DE 2, 3 Y 5
# Crear un arreglo de números enteros entre 0 y 150
numeros = list(range(151))

# Calcular los múltiplos de 2, 3 y 5 utilizando funciones estándar
multiplos_2_estandar = [x for x in numeros if x % 2 == 0]
multiplos_3_estandar = [x for x in numeros if x % 3 == 0]
multiplos_5_estandar = [x for x in numeros if x % 5 == 0]

# Calcular los múltiplos de 2, 3 y 5 utilizando funciones lambda
multiplos_2_lambda = (lambda x: [i for i in x if i % 2 == 0])(numeros)
multiplos_3_lambda = (lambda x: [i for i in x if i % 3 == 0])(numeros)
multiplos_5_lambda = (lambda x: [i for i in x if i % 5 == 0])(numeros)

print("Múltiplos de 2 (función estándar):", multiplos_2_estandar)
print("Múltiplos de 3 (función estándar):", multiplos_3_estandar)
print("Múltiplos de 5 (función estándar):", multiplos_5_estandar)

print("Múltiplos de 2 (función lambda):", multiplos_2_lambda)
print("Múltiplos de 3 (función lambda):", multiplos_3_lambda)
print("Múltiplos de 5 (función lambda):", multiplos_5_lambda)
```

3.- En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solución del problema de valor inicial:

\begin{equation}
y´=-y+x \hspace{3cm} y(-1)=5.
\end{equation}


Los estudiantes llegaron a la solución general cuya ecuación es:

\begin{equation}
y=(x-1)+C e^{-x}
\end{equation}

Apoye a sus compañeros y grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:

*   para valores de C > 0
*   cuando C = 0
*   para valores de C < 0

Señale en el gráfico las curvas que corresponden a cada uno de los tres casos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función que representa la solución general
def solucion_general(x, C):
    return (x - 1) + C * np.exp(-x)

# Valores de x para la gráfica
x_vals = np.linspace(-2, 3, 400)

# Graficar para valores de C > 0
for C_pos in [1, 2, 3]:
    y_vals_pos = solucion_general(x_vals, C_pos)
    plt.plot(x_vals, y_vals_pos, label=f'C > 0, C={C_pos}')

# Graficar cuando C = 0
y_vals_zero = solucion_general(x_vals, 0)
plt.plot(x_vals, y_vals_zero, label='C = 0', linestyle='--')

# Graficar para valores de C < 0
for C_neg in [-1, -2, -3]:
    y_vals_neg = solucion_general(x_vals, C_neg)
    plt.plot(x_vals, y_vals_neg, label=f'C < 0, C={C_neg}', linestyle='dashed')

# Establecer etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostrar el gráfico
plt.show()
```

4.- Sea la función:
\begin{equation}
\frac{3x + 7}{x^2 - 4}
\end{equation}

Utilizando la librería matplotlib:

1.- Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.

2.- Investigue en la literatura cómo graficar las asíntotas horizontales y verticales. Muéstrelas en el gráfico con líneas punteadas.

3.- Establezca límites en el eje "y" para evitar que la gráfica muestre valores muy altos o muy bajos cerca de las asíntotas verticales.

4.- Añada una leyenda indicando los elementos presentes en la gráfica.

```python
#Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return (3*x + 7) / (x**2 - 4)

# Generar valores de x en el rango especificado con más de 1000 muestras
x_vals = np.linspace(-10, 10, 1000)

# Evaluar la función en los valores de x
y_vals = f(x_vals)

# Graficar la función con línea continua
plt.plot(x_vals, y_vals, label='(3x+7)/(x^2-4)', color='blue')

# Establecer etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la Función (3x+7)/(x^2-4)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Mostrar el gráfico
plt.show()
```
