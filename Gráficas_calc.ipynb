```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.linspace(0.1, 2, 100)
s = np.linspace(0.1, 2, 100)
C, S = np.meshgrid(c, s)
p_levels = [0.5, 1, 1.5, 2]
for p in p_levels:
    P = C * S
    plt.contour(C, S, P, levels=[p], label=f'$p = {p}$')
plt.xlabel('$c$ (Flujo cardíaco)')
plt.ylabel('$s$ (Resistencia vascular sistemática)')
plt.title('Conjuntos de Nivel de Presión Sanguínea')
plt.legend()
plt.show()
```

```python
c = np.linspace(0.1, 2, 100)
s = np.linspace(0.1, 2, 100)
C, S = np.meshgrid(c, s)
p_levels = [0.5, 1, 1.5, 2]
for p in p_levels:
    P = C * S
    plt.contour(C, S, P, levels=[p], label=f'$p = {p}$')
point_a = {'c': 1, 's': 1.5}
point_b = {'c': 1.5, 's': 0.8}
plt.scatter(point_a['c'], point_a['s'], color='red', marker='o', label='A (Antes)')
plt.scatter(point_b['c'], point_b['s'], color='blue', marker='o', label='B (Después)')
plt.xlabel('$c$ (Flujo cardíaco)')
plt.ylabel('$s$ (Resistencia vascular sistemática)')
plt.title('Conjuntos de Nivel de Presión Sanguínea con Puntos A y B')
plt.legend()
plt.show()
```
