```python
n=input("Inserta un número del 1 al 10 para que la máquina lo adivine")
c=np.random.randint(0,10,size=1)
if c==n:
  print(f"tu número es: {c}")
else:
    d=np.random.randint(0,10,size=1)
    if d==n:
      print(f"tu número es: {d}")
    else:
      e=np.random.randint(0,10,size=1)
      if e==n:
        print(f"tu número es: {e}")
      else:
        f=np.random.randint(0,10,size=1)
        if f==n:
          print(f"tu número es: {f}")
        else:
          g=np.random.randint(0,10,size=1)
          if g==n:
            print(f"tu número es: {g}")
          else:
            h=np.random.randint(0,10,size=1)
            if h==n:
              print(f"tu número es: {h}")
            else:
              i=np.random.randint(0,10,size=1)
              if i==n:
                print(f"tu número es: {i}")
              else:
              j=np.random.randint(0,10,size=1)
                if j==n:
                  print(f"tu número es: {j}")
                else:
                k=np.random.randint(0,10,size=1)
                  if k==n:
                  print(f"tu número es: {k}")
                  else:
                    l=np.random.randint(0,10,size=1)
                    print(f"tu número es: {l}")
```
